```python
import os
import numpy as np
import cv2
import pickle
import time
import random
#import matplotlib.pyplot as plt
```

```python
DIR = r'C:\Users\Bhavya Goel\Downloads\arr_rec\arr_rec'
CAT = ['Right','Left']
```

```python
data = []
img_size=100
for cat in CAT:
    folder = os.path.join(DIR,cat)
    label = CAT.index(cat)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        im_arr = cv2.imread(img_path)
        im_arr = cv2.resize(im_arr,(img_size,img_size))
        data.append([im_arr, label])
```

```python
random.shuffle(data)
```

```python
X = []
y = []

for features, lables in data:
    X.append(features)
    y.append(lables)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X=X/255
```

```python
pickle.dump(X,open('X.pkl','wb'))
pickle.dump(y,open('y.pkl','wb'))
```
